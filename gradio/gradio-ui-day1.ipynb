```python
import gradio as gr
```

```python
def greet(input):
    print(f"Greet function called with '{input}' as input")
    return input.upper()

greet("Hello from gradio app!")
```

```python
gr.Interface(fn=greet, inputs="text", outputs="text", flagging_mode="never").launch()
```

```python
#Adding inbroweser flag to launch the app in the default web browser
gr.Interface(fn=greet, inputs="text", outputs="text", flagging_mode="never").launch(inbrowser=True)
```
